# QQ-Plot (Quantile Quantile Plot)

A QQ plot is a technique used to determine if two data sets come from populations that have the same distributions.

A QQ plot is a plot of the quantiles of the first data set against the quantiles of the second dataset. 

By quantiles we mean the fraction of points that lie within some certain values. At the 0.3 quantile, 30% of points lie below that point and 70% of the points are above.

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

normal_numbers = np.random.normal(0, 1, 1000)

fig = plt.figure()
res = stats.probplot(normal_numbers, plot=plt)
plt.show()
```
